### Aim
### **To implement a linear regression with regularization via gradient descent. to implement gradient descent with Lp norm, for 3 different values of p in (1,2]**

```python
import pandas as pd
```

```python
df = pd.read_csv("/content/sample_data/Boston.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.corr()
```

```python
X = df[["crim","zn","indus","nox","rm","age","dis","rad","tax","ptratio","black","lstat"]]
y = df["medv"]

from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

### Linear Regression


```python
from sklearn.linear_model import LinearRegression #importing all the required functions
regressor = LinearRegression()
"Syntax : varName = ModelName(modelHyperParams)"
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test) # given the features of the test dataset, it will give the final predictions
"Syntax : varName.predict(test_features)"
y_pred
```

```python
df1 = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df1.head()
```

```python
from sklearn import metrics 
print('R2- SCORE:', metrics.r2_score(y_test,y_pred))
```

### **Lasso Regression**

```python
from sklearn.linear_model import Lasso
```

```python
Lasso_Reg = Lasso() 
"Syntax : varName = ModelName(modelHyperParams)"
Lasso_Reg.fit(X_train, y_train)
```

```python
y_pred1 = Lasso_Reg.predict(X_test)
"Syntax : varName.predict(test_features)"
y_pred1
```

```python
df2 = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred1})
df2.head()
```

```python
from sklearn import metrics 
print('R2- SCORE:', metrics.r2_score(y_test,y_pred1))
```

### **Ridge Regression**

```python
from sklearn.linear_model import Ridge
Ridge_Reg = Ridge()
Ridge_Reg.fit(X_train, y_train)
```

```python
y_pred2 = Ridge_Reg.predict(X_test)
"Syntax : varName.predict(test_features)"
y_pred2
```

```python
df3 = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred2})
df3.head()
```

```python
from sklearn import metrics 
print('R2- SCORE:', metrics.r2_score(y_test,y_pred2))
```

### **CROSS-VALIDATION**

```python
from sklearn.model_selection import cross_validate
regressor = LinearRegression(normalize = True) #untrained model
```

```python
cv_results = cross_validate(regressor, X, y, cv=10, scoring = "r2")
cv_results['test_score'].mean()
```
